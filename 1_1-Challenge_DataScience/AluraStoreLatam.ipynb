### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
tiendas
```

```python
totalFacturado = tiendas['Precio'].sum()
print(f"Facturación total de todas las tiendas: {totalFacturado}")
```

```python
# Facturación por tienda
facTienda = tiendas.groupby('Lugar de Compra')['Precio'].sum()
print("\nFacturación por tienda:\n")
print(facTienda)
```

```python
# Total facturado por mes
tiendas ['Fecha de Compra'] = pd.to_datetime(tiendas['Fecha de Compra'])
facMes = round(tiendas.groupby(tiendas['Fecha de Compra'].dt.month)['Precio'].sum())
print("\nFacturación por mes:\n")
print(facMes)
```

```python
# venta promedio por casa tienda
promedioTienda = round(facTienda.mean())
print(f"\nVenta promedio por tienda: {promedioTienda}")
```

# 2. Ventas por categoría

```python
# Ventas por categoria
ventasCategoria = tiendas.groupby('Categoría del Producto')['Precio'].sum()
print("\nTotal de ventas por categoría:\n")
print(ventasCategoria)
```

# 3. Calificación promedio de la tienda


```python
calPromedioTienda = tiendas.groupby('Lugar de Compra')['Calificación'].mean().round(1)
print("\nCalificación promedio por tienda:\n")
print(calPromedioTienda)
```

# 4. Productos más y menos vendidos

```python
masVendidos = tiendas.groupby('Producto')['Producto'].count().sort_values(ascending=False)
print("\nLos 10 Productos más vendidos:\n")
print(masVendidos.head(10))
```

```python
menosVendido = tiendas.groupby('Producto')['Producto'].count().sort_values(ascending=True)
print("\nLos 10 Productos menos vendidos:\n")
print(menosVendido.head(10))
```

# 5. Envío promedio por tienda

```python
envioPromedio = round(tiendas.groupby('Lugar de Compra')['Costo de envío'].mean())
print("\nPromedio costo de envio por tienda:\n")
print(envioPromedio)
```
