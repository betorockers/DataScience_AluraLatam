##Mejorando el Feature Engineering

```python
visitas_ultima = df.groupby('fullVisitorId',as_index=False)
visitas_ultima = visitas_ultima['visitNumber'].max()
visitas_ultima.head()
```

```python
df.head()
```

```python
usuarios_visitas_unicos = df.drop_duplicates(subset=['fullVisitorId','visitNumber'])
usuarios_visitas_unicos.head()
```

```python
usuarios_visitas_unicos.shape
```

```python
visitas = pd.merge(visitas_ultima,usuarios_visitas_unicos,left_on=['fullVisitorId','visitNumber'],
                  right_on=['fullVisitorId','visitNumber'],how='left')
visitas.head()
```

```python
visitas.shape
```

```python
visitas_primera = df.groupby('fullVisitorId',as_index=False)
visitas_primera = visitas_primera['visitNumber'].min()
visitas_primera.head()
```

```python
visitas_primera.set_index('fullVisitorId',inplace=True)
visitas_primera.head()
```

```python
visitas = visitas.join(visitas_primera,how='left',on='fullVisitorId',rsuffix='_primera')
visitas.head()
```

# Aula05

##Separando la base de datos

```python
visitas.drop('fullVisitorId',axis=1,inplace=True)
y = visitas.transactionRevenue.copy()
X = visitas.drop('transactionRevenue',axis=1)
X.head()
```

```python
cuali = visitas.dtypes[visitas.dtypes == object].keys()
cuali
```

##Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
strings = list(X.operatingSystem_ultima.values.astype('str'))
```

```python
lbl = LabelEncoder()
```

```python
lbl.fit(strings)
```

```python
lbl.transform(strings)
```

```python
for columna in cuali:
    lbl = LabelEncoder()
    strings = list(X[columna].values.astype('str'))
    lbl.fit(strings)
    X[columna] = lbl.transform(strings)
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

##Regresi√≥n Lineal

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
reg_predict = reg.predict(X_test)
```

```python
reg_predict[reg_predict < 0] = 0
```

```python
resultados = pd.DataFrame()
resultados['revenue'] = y_test
resultados['predict'] = reg_predict
resultados['error'] = reg_predict - y_test
resultados.head()
```

```python
resultados[resultados.revenue > 0]
```

```python
np.sqrt(mean_squared_error(y_test,reg_predict))
```
