### **Parámetros de la función read_excel**

Documentación: https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html

```python
# exploraremos los parametros de read_excel para ver o revisar cualquier hoja en excel
percapita = pd.read_excel(archivo,sheet_name='emisiones_percapita')
```

```python
# las primeras 5 filas
percapita.head()
```

```python
# las ultimas 5 filas
percapita.tail()
```

```python
# cantidad de filas al azar
percapita.sample(5)
```

```python
# examinaremos la hoja llamada fuentes
fuentes = pd.read_excel(archivo,sheet_name='fuentes')
```

```python
fuentes.tail()
```

```python
# Examinares las hojas con los parametrtos nrows y usecols
# definiremos el intervalo de variables

intervalo = pd.read_excel(archivo,sheet_name='emisiones_C02', usecols ='A:D')
```

```python
intervalo
```

```python
intervalo2 = pd.read_excel(archivo,sheet_name='emisiones_C02', usecols ='A:D', nrows = 10)
```

```python
intervalo2
```
