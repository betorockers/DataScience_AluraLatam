### **1. Entendiendo el Problema**

```python
# importamos los archivos 
datos = pd.read_json('../Data/datos_hosting.json')
datos.head()
```

```python
# Normalizamos los datos importados para poder analizarlos mejor
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

```python
#datos.info()

```

### **2. datos numéricos**

```python
# extraeremos los registros de cada elemento en las columnas
columnas = list(datos.columns)
columnas
```

```python
# utilizaremos el metodo explode para revisar las listas desde la columna descripcion_local
data = datos.explode(columnas[3:])
data
```

```python
# dejaremos los indices en numeros no repetidos(recetearemos los indices)
data.reset_index(inplace = True, drop = True)
data.head()
```

```python
data.info()
```

```python
# tranformaremos los datos para que cada columna muestre el dato que tiene correctamente utilizaremos la biblioteca numpy(este metodo no funcionara si algun dato de la columna no es entero)
data['max_hospedes'] = data['max_hospedes'].astype(np.int64)
```

```python
# trasformamos las siguientes columnas que contienen numeros enteros
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
data[col_numericas] = data[col_numericas].astype(np.int64)
```

```python
# ahora trasnformaremos las columnas con numeros decimales
data['evaluacion_general'] = data['evaluacion_general'].astype(np.float64)
```

```python
# ahora trasnformaremos las columnas con numeros decimales y caracteres asociados($) utilizando el metodo apply() con una funcion lambda,si trabajo con series.replace(',', '').strip())
```

```python
data['precio'] = data['precio'].astype(np.float64)
```

```python
data['precio'].astype(np.float64)
```

```python
data.info()
```

```python
# ahora trasnformaremos las columnas con numeros decimales y caracteres asociados($, ',') utilizando el metodo applymap() con una funcion lambda,si trabajo con DataFrame
data[['cuota_deposito', 'cuota_limpieza']] = data[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x,: x.replace('$', '').replace(',', '').strip())
```

```python
# ahora si transformamos a tipo float64
data[['cuota_deposito', 'cuota_limpieza']] = data[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
data.info()
```

### **3.Datos de texto**

```python
data.head()
```

```python
data['descripcion_vecindad'] = data['descripcion_vecindad'].str.lower()
```

```python
# eliminaremos caracteres especiales que no utilizaremos, para esto usaremos str y regex(expresiones regulares)
data['descripcion_local'][3169]
```

```python
# utilizaremos regex
data['descripcion_local'] = data['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
```

```python
# Eliminaremos los guiones que no correspondan a palabras compuestas ejemplo 2-persons, pero si eliminamos el guion si se utiliza como separador one - two
data['descripcion_local'] = data['descripcion_local'].str.replace('(?<!\w)-(?>!\w)',' ',regex=True)
```
