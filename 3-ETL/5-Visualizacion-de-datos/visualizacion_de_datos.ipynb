# Visualizacion de datos con python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# nos enfocaremos en un grupo correspondiente por pais
df.set_index('Pais', inplace= True)
df
```

```python
# extraeremos info de un pais especifico
anos = list(map(str,range(1980,2014)))
anos
```

```python
# apliamos el filtro para ubicaras un pais, y obtener un pandas series con los años y la cantidad de inmigrantes por año
colombia = df.loc['Colombia', anos]
colombia
```

```python
chile = df.loc['Chile', anos]
chile
```

```python
# crearemos un dataFrame a traves de un diccionario
col_dict = {'año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()}
```

```python
# crearemos un dataFrame a traves de un diccionario
chl_dict = {'año': chile.index.tolist(),
            'Inmigrantes': chile.values.tolist()}
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
datos_chl = pd.DataFrame(chl_dict)
datos_chl.tail()
```

## Mostrar la informacion a traves de matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# grafico para Colombia 
plt.figure(figsize=(8,4))
plt.plot(datos_col['año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
# Agregar título y etiquetas a los ejes
plt.title('Inmigracion Colombiana hacia Canada')
plt.xlabel('Años', color='r')
plt.ylabel('Cantidad de Inmigrantes')
# Agregar la marca de agua (texto)
plt.text(0.5, 0.5, 'BetoRock Toledo', transform=plt.gca().transAxes,
          ha='center', va='center', color='y', fontsize=20, alpha=0.5)
plt.show()
```

```python
# grafico para Chile
plt.figure(figsize=(8,4))
plt.plot(datos_chl['año'], datos_chl['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
# Agregar título y etiquetas a los ejes
plt.title('Inmigracion Chilena hacia Canada')
plt.xlabel('Años', color='r')
plt.ylabel('Cantidad de Inmigrantes')
# Agregar la marca de agua (texto)
plt.text(0.5, 0.5, 'BetoRock Toledo', transform=plt.gca().transAxes,
          ha='center', va='center', color='#0000FF', fontsize=20, alpha=0.5)
plt.show()
```

## capitulo 2 **Creando figuras con Matplotlib**

```python
# Mejorando las graficas y el lienzo; fig(lienzo), ax(grafica), podemos obtener mas de una grafica 
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # esto hace lo mismo que xticks pero de forma automatica y dinamica
# titulos y leyendas
ax.set_title('Inmigracion de Colombianos hacia Canada\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Inmigrantes')
ax.text(0.5, 0.5, 'BetoRock Toledo', transform=plt.gca().transAxes,
          ha='center', va='center', color='#0000FF', fontsize=20, alpha=0.2)
plt.show()
```

```python
# Trabajaremos con 2 figuras esta vez

#! Grafica de la izquierda
fig, axs = plt.subplots(1,2, figsize=(10,3)) # el 1 indica la fila y el 2 indica las figuras a mostrar, figsize indica el tamaño del lienzo
axs[0].plot(datos_col['año'], datos_col['Inmigrantes']) # grafica de la izquierda
axs[0].set_title('Inmigracion de Colombianos hacia Canada\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Numero de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

#? Grafica de la derecha, grafico de cajas
axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigracion de Colombianos\nhacia Canada 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número dse Inmigrantes')
axs[1].grid()

plt.show()
```

Creando subplots en dos direcciones #1

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3) # espaciado entre graficos
fig.suptitle('Inmigracion de los 4 mayores paises sudamericanos\nhacia Canadá desde 1980 - 2013')

#grafico 1
axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')


#grafico 2
axs[0,1].plot(df.loc['Brasil', anos])
axs[0,1].set_title('Brasil')


#grafico 3
axs[1,0].plot(df.loc['Chile', anos])
axs[1,0].set_title('Chile')



#grafico 4
axs[1,1].plot(df.loc['Argentina', anos])
axs[1,1].set_title('Argentina')

# estableceremos un valor maximo para y, y un valor maximo para y
ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)


# ajustamos los años a un interlapso de 5 años para todos los graficos
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Año')
    ax.set_ylabel('Número de Inmigrantes')
    ax.grid()
    

plt.show()
```
